```python
import json
```

```python
# Load JSON data from a file
with open('dev_data.json', 'r') as f:
    data = json.load(f)

# Assuming the data is in a list format, access an element to retrieve the keys
sample_element = data[0]
columns = sample_element.keys()
print("Columns:", columns)

# # Check the number of rows
num_rows = len(data)
print("Number of rows:", num_rows)
```

```python
# Check the number of rows where the result is a win
win_count = 0
loss_count = 0
draw_count = 0
for row in data:
    if row['result'] == 'win':  # Compare with boolean value True instead of string 'true'
        win_count += 1
    if row['result'] == 'loss':
        loss_count += 1
    if row['result'] == 'draw':
        draw_count += 1
            
print(win_count)
print(loss_count)
print(draw_count)
```

```python
# Find null, NA, or blank values
null_values = []
for row in data:
    for key, value in row.items():
        if value is None or value == "" or value == "NA":
            null_values.append((key, value))

# Display null, NA, or blank values
for col, value in null_values:
    print("Null value in column:", col)
    print("Value:", value)
    print()
```

```python
# Count occurrences of user_id
user_id_counts = {}
for row in data:
    user_id = row['user_id']
    user_id_counts[user_id] = user_id_counts.get(user_id, 0) + 1

# Find user_ids with multiple occurrences
multiple_occurrences = {user_id: count for user_id, count in user_id_counts.items() if count > 1}

# Display user_ids with multiple occurrences
for user_id, count in multiple_occurrences.items():
    print("User ID:", user_id)
    print("Occurrences:", count)
    print()
```
