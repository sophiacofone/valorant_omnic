```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('logreg_csv_feature_results/df_coefs_logreg_postspike.csv')
# getting top 20
df_sorted = df.sort_values(by='abs_coef', ascending=False).head(20)
# semantic grouping
groups = {
    'Death-related': ['deaths', 'death','dead'],
    'Health-related': ['health', 'avg_health'],
    'Round-related': ['round_info','spike_planted'],
    'Ability-related': ['crowd_control','total_ability'],
    'Weapon-related': ['gun','melee','ammo'],
    'Elimination-related':['elims','damage'],
    'Economy-related':['credits'],

}
df['group'] = df['feature'].apply(lambda x: next((group for group, keywords in groups.items() if any(keyword in x for keyword in keywords)), 'Other'))
# Now continue with your plotting code
df_sorted = df.sort_values(by='abs_coef', ascending=False).head(20)
plt.figure(figsize=(10, 8))
colors = {'Death-related': 'red','Economy-related':'magenta', 'Health-related': 'green', 'Round-related': 'blue', 'Ability-related':'purple','Weapon-related':'yellow','Elimination-related':'cyan','Other': 'gray'}
plt.barh(df_sorted['feature'], df_sorted['abs_coef'], color=[colors[group] for group in df_sorted['group']])
plt.xlabel('Coefficient Value')
plt.ylabel('Feature')
plt.title('Top 20 Feature Importance with Semantic Grouping')
plt.gca().invert_yaxis()
plt.tight_layout()
# legend
legend_labels = [plt.Line2D([0], [0], color=color, label=group) for group, color in colors.items()]
plt.legend(handles=legend_labels)
plt.show()
```

```python
def vis(csv_in):
    df = pd.read_csv(csv_in)

    # Pivot the data for the heatmap
    heatmap_data = df(columns="feature", values="abs_coef")

    # Create the heatmap using seaborn
    plt.figure(figsize=(10, 8))
    sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", center=0)
    plt.title("Heatmap of Coefficients")
    plt.show()


vis('win_loss/logreg_csv_feature_results/df_coefs_logreg_attack.csv')
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz

# Load the dataset
data = load_iris()
X = data.data
y = data.target

# Train a decision tree model
clf = DecisionTreeClassifier()
clf.fit(X, y)

# Extract feature importance
feature_importance = clf.feature_importances_

# Create DOT data
dot_data = export_graphviz(
    clf,
    out_file=None,
    feature_names=data.feature_names,
    filled=True,
    rounded=True,
    special_characters=True,
    impurity=False,
)

# Visualize the decision tree using Graphviz
graph = graphviz.Source(dot_data)
graph.render("decision_tree")  # This will save the visualization as "decision_tree.pdf"
graph.view("decision_tree")    # This will open the visualization in your default PDF viewer
```
